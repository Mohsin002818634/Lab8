```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

olympic = pd.read_csv('olympic.csv')

print(olympic.head())
```

```python
fig, ax = plt.subplots()

ax.scatter(olympic['gold'], olympic['year'])

ax.set_title('Most Golds Won')
ax.set_xlabel('gold')
ax.set_ylabel('year')
```
